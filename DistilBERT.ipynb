```python
import pandas as pd
import numpy as np
import ktrain
from ktrain import text
import tensorflow as tf
```

```python
data_test = pd.read_csv('Corona_NLP_test.csv', dtype= str)
data_train = pd.read_excel('corona_NLP_train.xlsx', dtype = str)
```

```python
data_train.sample(5)
```

```python
text.print_text_classifiers()
```

```python
(train, val, preproc) = text.texts_from_df(train_df=data_train, text_column='OriginalTweet', label_columns='Sentiment',
                   val_df = data_test,
                   maxlen = 400,
                   preprocess_mode = 'distilbert')
```

```python
model = text.text_classifier(name = 'distilbert', train_data = train, preproc=preproc)
```

```python
learner = ktrain.get_learner(model = model,
                             train_data = train,
                             val_data = val,
                             batch_size = 6)
```

```python
learner.fit_onecycle(lr = 2e-5, epochs=2)
```

```python
predictor = ktrain.get_predictor(learner.model, preproc)
```
